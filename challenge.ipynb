```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
        
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
# cria um dataframe para o faturamento
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)

df_faturamento
```

```python
# calcula a média do faturamento total

media_faturamento = df_faturamento.valor.mean()

media_faturamento
```

```python
df_faturamento.data_ref = pd.to_datetime(df_faturamento['data_ref']).dt.strftime('%m')
```

```python
# remove a formatação científica do gráfico
formatter = ticker.ScalarFormatter()
formatter.set_scientific(False)
plt.gca().yaxis.set_major_formatter(formatter)

# Cria o gráfico e define os nomes
plt.bar(df_faturamento.data_ref, df_faturamento.valor)
plt.ylabel('Faturamento mensal')
plt.xlabel('Mês')
plt.title('Gráfico vertical')
```
